```python
df1=df.iloc[:,:10]            #Selecting columns only with non null values
```

```python
#As we can see many columns names are just questions ,will short them 
```

```python
df1.columns = df1.columns.str.replace("'"," ")
```

```python
df1
```

```python
df1=df1.rename(columns={'Do you eat Eggs/ Non-Veg?':'Non_veg',
                        'What s your favourite beverage?':'Fav_drink',
                        'What would you prefer?':'prefer1',
                        'What would you prefer?.1':'prefer2',
                        'What would you prefer?.2':'prefer3',
                        'What would you prefer?.3':'prefer4' })
df1.head()                                                      #Renaming the columns for convenience
```

```python
#Now selecting the data for predictor and target variables as X is predictor and Y is target variables
```

```python
X=df1.iloc[:,1:6]
X                            #predictor data
```

```python
Y=df1.iloc[:,6:]
Y             #target data
```

```python
Y.values
```

```python
# we can create groups of age from age column
```

```python
X.Age.min(),X.Age.max()
```

```python
age_group=[]
for i in X.Age:
    if(i>12 and i<=25):
        age_group.append('12-25')
    elif (i>25 and i<=35):
        age_group.append('25-35')
    else:
        age_group.append('35-50')
    
```

```python
sb.countplot(X.Age)
```

```python
X=X.drop('Age',axis=1)
X['age_group']=age_group
X
```

```python
X.Non_veg=X.Non_veg.astype(str)                    #veg_non_veg column is in boolean form convert it into object form
```

```python
X1=pd.get_dummies(X)
X1
```

```python
#Splitting data for training and testing of model
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X1,Y,test_size=0.2,random_state=21)
```

```python
#Creating machine learning models
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
from sklearn.multioutput import MultiOutputClassifier
#Adaboost
clf=MultiOutputClassifier(AdaBoostClassifier(DecisionTreeClassifier(random_state=42,max_depth=80),n_estimators=500))
```

```python
#Randomforest
clf=MultiOutputClassifier(RandomForestClassifier(n_estimators=500,random_state=42,max_depth=200))
```

```python
#Decisiontree
clf=MultiOutputClassifier(DecisionTreeClassifier(random_state=42,max_depth=500))
```

```python
clf.fit(xtrain,ytrain)
clf.score(xtrain,ytrain)
```

```python
pred=clf.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix,classification_report ,accuracy_score
```

```python
y=pred[:,0]
y.shape
```

```python
y_2=pred[:,1]
y_2.shape
```

```python
#convert into array
y_arr=np.array(ytest)
```

```python
auc_y1 = confusion_matrix(y_arr[:,0],y)
auc_y2 = confusion_matrix(y_arr[:,1],y_2)
```

```python
auc_y1,auc_y2
```

```python
cl1=classification_report(y_arr[:,0],y)
cl2=classification_report(y_arr[:,1],y_2)
```
